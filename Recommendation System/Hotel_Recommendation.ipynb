# Imports

```python
import pandas as pd
import numpy as np
from textblob import TextBlob
from sklearn.metrics.pairwise import cosine_similarity
```

# Load .csv file and data cleaning

```python
df = pd.read_csv("Hotel_Reviews.csv")

df = df.rename(columns = {'reviews.text': 'review', 'reviews.title': 'title','reviews.rating': 'rating','reviews.username':'username'}, inplace = False)
df.drop(['categories', 'city','country','longitude','postalCode','reviews.date','reviews.dateAdded','reviews.doRecommend','reviews.id','reviews.userCity','reviews.userProvince'], axis = 1,inplace = True)
df=df[df["name"].str.contains("�")==False]
df.drop_duplicates(subset ='review',keep = False, inplace = True)
df
```

# Find Sentiment on Text(Reviews)

```python
def sentiment_calc(review):
    try:
        return TextBlob(str(review)).sentiment.polarity
    except:
        return None
df['sentiment'] = df['review'].apply(sentiment_calc)
df
```

# Apply Multiplication B/W Rating&Sentiment

```python
df['Updated_score'] = df['rating']*df['sentiment']
df
```

# Make Username into Normal Form

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['username'] = le.fit_transform(df['username'])
```

# Make function for classify updated_score

```python
# dict1 = {'1':[-5,-4,-3.75,-3,-2.5,-2.25,-2,-1.5,-1.25,-1,-0.75,-0.5,-0.25],'2':[-0.24,0.25,0.5,0.75,1] ,'3': [1.01,1.25,1.5,2],'4': [2.01,2.25,2.5,3],'5':[3.01,3.75,4,5]}

def fun(val):
    if val >= -5 and val <=-3:
        return 1
    
    if val >= -3  and val <= -1:
        return 2
    
    if val >= -1 and val <= 1 :
        return 3
    
    if val >= 1 and val <= 3:
        return 4
    if val >= 3  and val <=5:
        return 5
    
    
```

# Apply function on updated_score and put into New_score column

```python
df['New_score'] = df['Updated_score'].apply(fun)
df['New_score'] = pd.to_numeric(df['New_score'])
df
```

# Pivot table of name, username and New_score

```python
df_pivot = df.pivot_table(index='name',columns='username',values='New_score').fillna(0)
df_pivot
```

# Sparse Matrix (Compressed Sparse Row)

```python
from scipy.sparse import csr_matrix

df_pivot_matrix = csr_matrix(df_pivot.values) 
print(df_pivot_matrix)
```

# Fitting data into NearestNeighborsModel

```python
from sklearn.neighbors import NearestNeighbors

model_knn = NearestNeighbors(metric = 'cosine', n_neighbors=30, radius=1)
model_knn.fit(df_pivot_matrix)
```

# Matrix of cosine similarity

```python
similarity_matrix = cosine_similarity(df_pivot)
similarity_matrix
```

```python
data = list(df_pivot.index) #shows list of Hotels in data-set
print(data)
```

```python
name = (input('Enter Hotel Name according to data set : '))
print(name)
```

```python
query_index = data.index(name) #shows index of productID by USER
print(query_index)
```

```python
similarity, indices = model_knn.kneighbors(df_pivot.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 8)
print(similarity) #shows similarity distance through productID by USER
print(indices) #shows indexs of productID by USER
```

# Create dictionary and store recommendations of Hotel Name which is given by USER

```python
data_dict={}
for i in range(0, len(similarity.flatten())):    #gives length of similarity array

    if i == 0:
        print('Recommendations for {0}:\n'.format(df_pivot.index[query_index]))
    else:
        data_dict[str(df_pivot.index[indices.flatten()[i]])] = float(similarity.flatten()[i])
        print(f'\n{df_pivot.index[indices.flatten()[i]]}, is similarity distance = with {similarity.flatten()[i]:.20f}:')

print(data_dict)
```
